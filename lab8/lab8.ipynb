**Тезис 1**
В Play Market больше всего приложений жанра "game"



```python
games = pd.read_csv('playmarket.csv')

```

```python
Category = games.groupby('Category').size().plot(kind='bar')
Category.set_title('Классификация приложений Google Play Store')
Category.set_ylabel('Количество приложений')
Category.set_xlabel('Категория приложения')
plt.show()
```

Тезис опровергнут. Большая часть приложений категории "Family"
